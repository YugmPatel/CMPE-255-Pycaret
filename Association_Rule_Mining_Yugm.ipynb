```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

# Load your dataset
data = pd.read_csv('market.csv', delimiter=';') # Specify the separator as semicolon
data.head()
```

```python
transactions = data.groupby('Bread')['Honey'].apply(list).reset_index()
```

```python
transactions_list = transactions['Honey'].tolist()
te = TransactionEncoder()
```

```python
te_ary = te.fit(transactions_list).transform(transactions_list)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
df_encoded.head()
```

```python
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules.head()
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.55)
rules.head()
```

```python
import matplotlib.pyplot as plt

# Plot lift vs confidence
plt.scatter(rules['confidence'], rules['lift'], alpha=0.25)
plt.title('Association Rules')
plt.xlabel('Confidence')
plt.ylabel('Lift')
plt.show()
```
